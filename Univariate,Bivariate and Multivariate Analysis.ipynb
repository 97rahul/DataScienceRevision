```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
```

```python
dataset = load_wine()
```

```python
dataset
```

```python
X = pd.DataFrame(dataset['data'],columns = dataset['feature_names'])
```

```python
X
```

```python
Y = pd.DataFrame(dataset.target,columns = ['target'])
```

```python
Y
```

```python
data = pd.concat([X,Y],axis = 1)
```

## Univariate Analysis

```python
data_0 = data.loc[data['target'] == 0]
data_1 = data.loc[data['target'] == 1]
data_2 = data.loc[data['target'] == 2]
```

```python
plt.plot(data_0['alcalinity_of_ash'],np.zeros_like(data_0["magnesium"]),'o' )
plt.plot(data_1['alcalinity_of_ash'],np.zeros_like(data_1["magnesium"]),'o' )
plt.plot(data_2['alcalinity_of_ash'],np.zeros_like(data_2["magnesium"]),'o' )
```

## Bivariate Analysis

```python
sns.FacetGrid(data,hue = 'target',size = 5).map(plt.scatter,'alcalinity_of_ash','alcohol').add_legend()
```

## MultiVariate Analysis

```python
sns.pairplot(data,hue = "target",size = 3)
```
